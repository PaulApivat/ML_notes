```python
import pandas as pd

gender_submission = pd.read_csv("gender_submission.csv")
gender_submission.head()
```

```python
train_data = pd.read_csv("train.csv")
train_data.describe()
```

```python
test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
# filter Train.csv data, for Sex == 'female', then select Survived columns

women = train_data.loc[train_data.Sex == 'female']['Survived']
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

```python
men = train_data.loc[train_data.Sex == 'male']['Survived']
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
# Build Decision Tree (review)
# Then RandomForest (getting-started-with-titanic, pt 3)
# Try 5-fold cross validation (Ch 3 ML-book & intermediate ml)
# Calculate: (lectures)
# - recall
# - precision
# - F-measure
# - Average F-measure
```

```python
# Create target object (train_y) - survival (binary)

train_y = train_data.Survived
train_y
```

```python
# Create train_X (features in the training data)

# Pclass = Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd); proxy for Socio-economic Status
# Sibsp = # of siblings / spouses aboard the Titanic
# Parch = # of parents / children aboard the Titanic
# fare = Passenger fare

features = ['Pclass', 'Sex', 'SibSp', 'Parch']
print(features)
train_X = pd.get_dummies(train_data[features])
print(train_X)
train_X.head()
```

```python
# Create X_test (features in the testing data)

X_test = pd.get_dummies(test_data[features])
print(X_test)
```

```python
# Data Already Split into Validation & Training
# no need to do train_X, X_test, train_y, y_test = train_test_split(X, y, random_state=1)

from sklearn.tree import DecisionTreeClassifier

# Specify Model
tree_clf = DecisionTreeClassifier(max_depth=3, random_state=1)

# Fit Model
tree_clf.fit(train_X, train_y)

# Prediction (on Testing data)
predictions = tree_clf.predict(X_test)

# Output
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
print(output)

# Save Output
output.to_csv("my_tree_submission.csv", index=False)
```

```python
# Calculate mean absolute error (load test.csv, see above)

from sklearn.metrics import mean_absolute_error

# See Output from previous
y_test = output.Survived

test_mae = mean_absolute_error(predictions, y_test)
print("Validation MAE for Decision Tree Classifier:", test_mae)
```
